```
import os
import shutil
```

```
pasta_pega = '/content/drive/My Drive/ColabNotebooks/Projetos Portifolio/Monkey Species Prediction/dados/training'
pasta_salva = '/content/drive/My Drive/ColabNotebooks/Projetos Portifolio/Monkey Species Prediction/dados/testing'
```

```
for i in os.listdir(pasta_pega):
  pasta = os.listdir(pasta_pega+'/'+i)
  print(len(pasta))
```

```
for i in os.listdir(pasta_pega):

  if os.path.isdir(pasta_salva+'/'+i) == False:
    os.makedirs(pasta_salva+'/'+i)

  imagens = os.listdir(pasta_pega+'/'+i)[:10]
  
  for j in imagens:
    imagem_pega = pasta_pega+'/'+i+'/'+j
    imagem_salva = pasta_salva+'/'+i
    shutil.move(imagem_pega, imagem_salva)
    print(imagem_salva+'/'+j)
```

```
for i in os.listdir(pasta_pega):
  pasta = os.listdir(pasta_pega+'/'+i)
  print(len(pasta))
```
